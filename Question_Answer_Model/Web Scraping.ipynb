### Get Question answer from ssa.gov website

```python
url_ssn = "https://faq.ssa.gov/en-US/"
```

```python
ctx = ssl.create_default_context()
ctx.check_hostname= False
ctx.verify_mode = ssl.CERT_NONE
```

```python
req = Request(url_ssn, headers={'User-Agent' : 'Mozilla/5.0'}) 
```

```python
webpage = urlopen(req).read()
soup = BeautifulSoup(webpage, 'html.parser')
```

```python
print(soup.prettify())
```

```python
ssn_QA = soup.find("article")
```

```python
ssn_QA
```

```python
rows  = ssn_QA.find_all('a')
```

```python
rows[0].contents[0]
```

```python
rows
```

```python
urls = {}
base_url = "https://faq.ssa.gov"
for row in rows:
    title = row.contents[0]
    if(row.get('href') != None and title != "\n"):
        link = base_url + row.get('href')
        urls[title] = link
```

```python
urls
```

```python
questions = []
Answers = []
for key,value in urls.items():
    ctx = ssl.create_default_context()
    ctx.check_hostname= False
    ctx.verify_mode = ssl.CERT_NONE
    req = Request(value, headers={'User-Agent' : 'Mozilla/5.0'}) 
    webpage = urlopen(req).read()
    soup = BeautifulSoup(webpage, 'html.parser')
    answers = soup.find("script", {"type":"application/ld+json"})
    answers.contents[0] = answers.contents[0].replace('\r\n ', '')
    mainEntity= json.loads(answers.contents[0], strict=False)["mainEntity"]
    answer = mainEntity[0]["acceptedAnswer"]['text']
    Answers.append(answer)
    questions.append("Q: " + key)
```

```python
import pandas as pd 
```

```python
d = pd.DataFrame({"Questions" : questions , "Answers" : Answers})
```
